```python
import os
import cv2
import matplotlib.pyplot as plt

# Specify the relative path to your image
relative_image_path = 'C:/Users/ksrin/OneDrive/Documents/plant_disease_detection/test/Apple___healthy.JPG'

# Get the absolute path
absolute_image_path = os.path.abspath(relative_image_path)

# Print the absolute image path to verify it's correct
print(f"Absolute image path: {absolute_image_path}")

# Check if the file exists at the specified path
if os.path.exists(absolute_image_path):
    # Reading the image in default mode
    img = cv2.imread(absolute_image_path)

    # Check if the image is loaded successfully
    if img is None:
        print("Error: Image not loaded. The image file might be corrupted.")
    else:
        # Convert the image from BGR to RGB
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        # Display the image
        plt.imshow(img)
        plt.axis('off')  # Hide the axes
        plt.show()
else:
    print(f"Error: The path {absolute_image_path} does not exist. Check the file path and ensure the image exists.")
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np

# Correct the path if necessary
corrected_image_path = 'C:/Users/ksrin/OneDrive/Documents/plant_disease_detection/test/Apple___healthy.JPG'

# Print the corrected image path to verify it's correct
print(f"Corrected image path: {corrected_image_path}")

# Check if the file exists at the specified path
if os.path.exists(corrected_image_path):
    # Load and preprocess the image
    image = load_img(corrected_image_path, target_size=(128, 128))
    input_arr = img_to_array(image)
    input_arr = np.array([input_arr])  # Convert single image to a batch.
    predictions = model.predict(input_arr)
    print(predictions)
else:
    print(f"Error: The path {corrected_image_path} does not exist. Check the file path and ensure the image exists.")
```

```python
print(predictions)
```

```python
result_index = np.argmax(predictions) #Return index of max element
print(result_index)
```
